```python
#библиотеки для обработки и визуализации информации
import pandas as pd
import seaborn as sns
from scipy.stats import kurtosis
from scipy.stats import skew
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#библиотеки для машинного обучения
from sklearn.neighbors import KNeighborsClassifier       #метод К-ближайших соседей
from sklearn.naive_bayes import GaussianNB               #наивный байесовский классификатор
from sklearn.tree import DecisionTreeClassifier          #деревья решений
from sklearn.svm import SVC                              #метод опорных векторов
from sklearn.linear_model import LogisticRegression      #логистическая регрессия
from xgboost import XGBClassifier                        #бустинг
from sklearn.ensemble import (GradientBoostingClassifier, #ансамбли
                             RandomForestClassifier,
                             ExtraTreesClassifier)
import xgboost as xgb

#метрики качества моделей
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

#вспомогательный для машшинного обучения
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold)

#функция для просмотра категориальных фичей (одна фича)
def info_categorial(pd_series):
    labels = pd_series.index.values
    sizes = [i for i in pd_series]
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
    plt.show()
    print(pd_series)

#функция для просмотра категориальных фичей (несколько фичей)
def info_categorial_2(list_pd_series, data, col):
    row = int((len(list_pd_series))/col)+1
    fig = plt.figure(figsize=(15,15))
    for l, i in enumerate(list_pd_series):
        ax = fig.add_subplot(row, col, l+1)
        pd_series = data[i].groupby(data[i]).count()
        labels = pd_series.index.values
        sizes = [k for k in pd_series]
        ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
        ax.set_title(pd_series.name)
    plt.show()

#функция для просмотра категориальных фичей (несколько фичей - диаграмма слева неотключенные, справа - отключенные)
def info_categorial_3(list_pd_series, data):
    row = len(list_pd_series)
    fig = plt.figure(figsize=(20,60))
    for l, i in enumerate(list_pd_series):
        
        ax = fig.add_subplot(row, 2, l+l+1)
        pd_series1 = data[data.Churn=='No'][i].groupby(data[i]).count()
        labels1 = pd_series1.index.values
        sizes1 = [k for k in pd_series1]
        ax.pie(sizes1, labels=labels1, autopct='%1.1f%%', shadow=True, startangle=90)
        ax.set_title(pd_series1.name)
        
        ax = fig.add_subplot(row, 2, l+l+2)
        pd_series2 = data[data.Churn=='Yes'][i].groupby(data[i]).count()
        labels2 = pd_series2.index.values
        sizes2 = [k for k in pd_series2]
        ax.pie(sizes2, labels=labels2, autopct='%1.1f%%', shadow=True, startangle=90)
        ax.set_title(pd_series2.name)        
    plt.show()

#функция для просмотра категориальных фичей (несколько фичей - диаграмма слева неотключенные, справа - отключенные)
#по группам GroupMonthlyCharges
def info_categorial_4(list_pd_series, data, group_charges):
    row = len(list_pd_series)
    fig = plt.figure(figsize=(20,60))
    for l, i in enumerate(list_pd_series):
        
        ax = fig.add_subplot(row, 2, l+l+1)
        pd_series1 = data[(data.Churn=='No') & (data.GroupMonthlyCharges==group_charges)][i].groupby(data[i]).count()
        labels1 = pd_series1.index.values
        sizes1 = [k for k in pd_series1]
        ax.pie(sizes1, labels=labels1, autopct='%1.1f%%', shadow=True, startangle=90)
        ax.set_title(pd_series1.name)
        
        ax = fig.add_subplot(row, 2, l+l+2)
        pd_series2 = data[(data.Churn=='Yes') & (data.GroupMonthlyCharges==group_charges)][i].groupby(data[i]).count()
        labels2 = pd_series2.index.values
        sizes2 = [k for k in pd_series2]
        ax.pie(sizes2, labels=labels2, autopct='%1.1f%%', shadow=True, startangle=90)
        ax.set_title(pd_series2.name)
        
    plt.show()
    
#функция для расчета статистик "числовых" фичей
def my_basic_research(series, bins_ = 25):
    print(f'Базовые метрики для: {series.name}')
    print(series.describe())
    print("------------------------------------")
    
    print("Самые популярные значения метрики, топ 5")
    print(series.value_counts().nlargest(5))
    print("------------------------------------")
    
    print("Эксцесс ", kurtosis(series))
    print("Ассиметрия ", skew(series))
    plt.subplot(121)
    series.hist(bins=bins_)
    plt.title(f'Распределение данных')
    plt.subplot(122)
    plt.boxplot(series, showmeans=True)
    plt.title(f'Ящик с усами')

#функция для подбора параметров моделей
def gridsearch_for_coursework(clf, param, cv, optim_score, Xtr, Ytr, Xts, Yts):
    GridSearch_model = GridSearchCV(estimator = clf, param_grid = param, cv=cv,
                                    scoring=optim_score, n_jobs = -1).fit(Xtr, Ytr)
    
    tn, fp, fn, tp = confusion_matrix(Yts, GridSearch_model.predict(Xts)).ravel()
    precision_ = precision_score(Yts, GridSearch_model.predict(Xts))
    recall_ = recall_score(Yts, GridSearch_model.predict(Xts))
    stacking_AUC_ = roc_auc_score(Yts, GridSearch_model.predict(Xts))
    f1_score_ = f1_score(Yts, GridSearch_model.predict(Xts))
    
    print(f'tn: {tn:.2f}, fp: {fp:.2f}, fn: {fn:.2f}, tp: {tp:.2f}')
    print(f'precision_score: {precision_:.2f}')
    print(f'recall_score: {recall_:.2f}')
    print(f'Stacking AUC: {stacking_AUC_:.2f}')
    print(f'f1_score: {f1_score_:.2f}')
    
    metrics = {'TN': tn, 'FP': fp, 'FN': fn, 'TP': tp, 'Precision': precision_, 'Recall': recall_,
               'Stacking AUC': stacking_AUC_, 'F1_score': f1_score_}
    return GridSearch_model, metrics

#функция бустинг
def boost(x_tr, y_tr, x_ts, y_ts, params, num_rounds, probability=0.5):
    dtrain = xgb.DMatrix(x_tr, y_tr)
    dtest = xgb.DMatrix(x_ts, y_ts)
    xgb_model = xgb.train(params, dtrain, num_rounds)
    xgb_predict = xgb_model.predict(dtest)
    
    tn, fp, fn, tp = confusion_matrix(y_ts, xgb_predict>probability).ravel()
    precision_ = precision_score(y_ts, xgb_predict>probability)
    recall_ = recall_score(y_ts, xgb_predict>probability)
    stacking_AUC_ = roc_auc_score(y_ts, xgb_predict>probability)
    f1_score_ = f1_score(y_ts, xgb_predict>probability)
    
    print(f'tn: {tn:.2f}, fp: {fp:.2f}, fn: {fn:.2f}, tp: {tp:.2f}')
    print(f'precision_score: {precision_:.2f}')
    print(f'recall_score: {recall_:.2f}')
    print(f'Stacking AUC: {stacking_AUC_:.2f}')
    print(f'f1_score: {f1_score_:.2f}')
    
    metrics = {'TN': tn, 'FP': fp, 'FN': fn, 'TP': tp, 'Precision': precision_, 'Recall': recall_,
               'Stacking AUC': stacking_AUC_, 'F1_score': f1_score_}
    return xgb_model, metrics 

# функция стекинга
# from sklearn.model_selection import cross_val_predict
# def stacking(models, meta_alg, data_train, targets_train, data_test, targets_test=None, test_size=None, cv=5):
#     if test_size is None:
#         meta_mtrx = np.empty((data_train.shape[0], len(models)))        
        
#         for n, model in enumerate(models):
#             meta_mtrx[:, n] = cross_val_predict(model, data_train, targets_train, cv=cv, method='predict')
#             model.fit(data_train, targets_train)
            
#         meta_alg.fit(meta_mtrx, targets_train)
        
#         meta_mtrx_test = np.empty((data_test.shape[0], len(models)))
        
#         for n, model in enumerate(models):
#             meta_mtrx_test[:, n] = model.predict(data_test)
            
#         meta_predict = meta.predict(meta_mtrx_test)
        
#         if targets_test is not None:
#             tn, fp, fn, tp = confusion_matrix(targets_test, meta_predict).ravel()
#             print(f'tn: {tn:.2f}, fp: {fp:.2f}, fn: {fn:.2f}, tp: {tp:.2f}')
#             print(f'precision_score: {precision_score(targets_test, meta_predict):.2f}')
#             print(f'recall_score: {recall_score(targets_test, meta_predict):.2f}')
#             print(f'Stacking AUC: {roc_auc_score(targets_test, meta_predict):.2f}')
#             print(f'f1_score: {f1_score(targets_test, meta_predict):.2f}')
    
#     elif test_size > 0 and test_size < 1:
#         train, valid, train_true, valid_true = train_test_split(data_train, 
#                                                                 targets_train,
#                                                                 test_size=test_size)
        
#         meta_mtrx = np.empty((valid.shape[0], len(models)))
        
#         for n, model in enumerate(models):
#             model.fit(train, train_true)
#             meta_mtrx[:, n] = model.predict(valid)
            
#         meta_alg.fit(meta_mtrx, valid_true)
        
#         meta_mtrx_test = np.empty((x_test.shape[0], len(models)))
        
#         for n, model in enumerate(models):
#             meta_mtrx_test[:, n] = model.predict(data_test)
            
#         meta_predict = meta.predict(meta_mtrx_test)
        
#         if targets_test is not None:
#             tn, fp, fn, tp = confusion_matrix(targets_test, meta_predict).ravel()
#             print(f'tn: {tn:.2f}, fp: {fp:.2f}, fn: {fn:.2f}, tp: {tp:.2f}')
#             print(f'precision_score: {precision_score(targets_test, meta_predict):.2f}')
#             print(f'recall_score: {recall_score(targets_test, meta_predict):.2f}')
#             print(f'Stacking AUC: {roc_auc_score(targets_test, meta_predict):.2f}')
#             print(f'f1_score: {f1_score(targets_test, meta_predict):.2f}')
        
#     else:
#         raise ValueError("test_size must be between 0 and 1")
        
#     return meta_predict
```

*Для обучения будем испытывать несколько моделей: К-блиайших соседей, дерево решений, случайный лес, градиентный бустинг, ExtraTreesClassifier, Логистическая регрессия, Наивный Байес, XGBoost. Определим параметры, которые мы будем подбирать с помощью кросс-валидации*

```python
knn_params = {'n_neighbors': np.arange(1,11,1),                            
              'weights': ['uniform', 'distance'],
              'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
              'leaf_size': np.arange(10,50,10)
}

tree_params = {'criterion': ['gini', 'entropy'],
               'splitter': ['best', 'random'],
               'max_features': ['auto', 'sqrt', 'log2']
}

rfc_params = {'n_estimators': range(10, 100, 10),
              'min_samples_leaf': range(1, 5)
}

# svc_params = {'degree': np.arange(1,8,1),
#               'C': np.arange(0.1, 1, 0.2),
#               'shrinking': [True, False]
# }

gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)
}

et_params = {'n_estimators':range(10, 100, 10),
    'max_depth': range(1, 15),
}

lr_params = {'C': np.arange(0.1, 1, 0.1),
             'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
}

bay_params = {'priors': [None],
}

xgb_params = {'n_estimators': np.arange(1,20,1),
            'learning_rate': np.arange(0.05, 0.5, 0.05),
            'max_depth': np.arange(2, 10, 1)
}

optim_score = 'roc_auc'
```

Также нам важно не ошибиться в отнесении клиента к классу неотток (целевая переменная со значением 0 или False), т.е. хуже потерять клиента (фактическое значение целевой переменной 1, а предсказанное 0), которого модель определила как "неотток", а вместе с ним и доход для компании, чем мы ошибочно причислим клиента в группу отток, а он останется в компании и продолжит приносить доход. Таким образом, нам важно оценить модели по метрикам False Negative и связанную с ним recall (полнота)*

*Посмотрим на эти метрики*

```python
fig = plt.figure(figsize=(15,5))

ax = fig.add_subplot(1, 2, 1)
df_score['FN'].sort_values().plot(kind='bar').legend(bbox_to_anchor=(0.2, 1))

ax = fig.add_subplot(1, 2, 2)
df_score[list_score[1]].sort_values(ascending=False).plot(kind='bar').legend(bbox_to_anchor=(0.95, 1))
```

*Но метрика точности тоже имеет значение, поскольку компания несет определенные расходы на удержание клиентов, поэтому нам важно также меньше ошибаться в случае отнесения клиентов в группу оттока. Иначе компания зря несет издержки на клиентов, которых модель причисляла к оттоку, а они на самом деле остаются лояльными к услугам компании. Таким образом, оценим модели по метрикам False Positive и связанную с ним precision (точность)*
